#This is highly Imbalanced Dataset

0 : Legit Transactions

1 : Fraudulent Transactions

```python
# separating the legit and fraudulent transactions

# save credit card data with the 'class' value is 0 to legit variable
legit = credit_card_data[credit_card_data.Class == 0]

fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

#Under-sampling

Build a sample dataset containing similar distribution of Legit & Fraudulent Transactions

Number of Fraudulent Transactions : 492

```python
# take 492 random sampling value from legit dataframe
legit_sample = legit.sample(n=492)
```

```python
print(legit_sample.shape)
```

#Concatenate the Two DataFrames

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
# axis=0 means concatenate after the last row of legit_sample (there is no additional column)
# axis=1 means concatenate after the last column of legit_sample (fraud column, there is no additional row)
```
